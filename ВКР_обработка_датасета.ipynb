```python
df_without_delivery = df[df.Группа2.isnull()==False]#без услуг
df_without_delivery.info()
```

```python
nan_indexes_Регион = df_without_delivery[df_without_delivery.Регион.isnull()].index #заполинили пустоты в регионах
df_without_delivery.loc[nan_indexes_Регион, 'Регион'] = pd.Series(df_without_delivery['ГородМагазина'].loc[nan_indexes_Регион], index=nan_indexes_Регион)
```

```python
df_without_delivery['Регион'].value_counts()
```

```python
df_without_delivery.Регион.replace('0', np.nan, inplace=True) #убирает нулевой регион
df_without_delivery.Регион.replace('nan', np.nan, inplace=True) #убирает нулевой регион
df_without_delivery.ГородМагазина.replace('0', np.nan, inplace=True) #убирает нулевой магазин
```

```python
nan_price = df_without_delivery[df_without_delivery.ЦенаЗакупки.isnull()]['ID_SKU'].unique()

for i in range(len(nan_price)):
    arr_i = np.sort(df_without_delivery[df_without_delivery.ID_SKU==nan_price[i]]['ЦенаЗакупки'].unique())
    if len(arr_i)>1: #заполнили пропуски на продуктах, где нашлась хоть одна цена, а это почти 700 строк
        a = df_without_delivery[df_without_delivery.ID_SKU==nan_price[i]].copy()
        cname = 'ЦенаЗакупки'
        prev_value = None
        for i, row in a.iterrows():
            if not np.isnan(row[cname]):
                prev_value = row[cname]
                break
        assert not prev_value is None

        for i, row in a.iterrows():
            if np.isnan(row[cname]):
                df_without_delivery.at[i, cname] = prev_value
            else:
                prev_value = row[cname]
```

```python
a = df_without_delivery[df_without_delivery.Маржа.isnull()]#заполнили пропуски в маржа
for i, row in a.iterrows():
    if (not np.isnan(row['Цена']))&(not np.isnan(row['ЦенаЗакупки'])):
        df_without_delivery.at[i, 'Маржа'] = row['Цена'] - row['ЦенаЗакупки']
```

```python
arr_nan_ = df_without_delivery[df_without_delivery.Маржа.isnull()].НомерЗаказаНаСайте.unique()
df_without_delivery.drop(df_without_delivery[df_without_delivery.НомерЗаказаНаСайте.isin(arr_nan_)].index, inplace=True) #удаляем чеки без ценызакупки
df_without_delivery.reset_index(drop=True)
```

## Результаты по таблице без доставки (Размещенные товары)

```python
def summary_all(x):
    result = {
        'Выручка': x['СуммаСтроки'].sum(),
        'Абсолютная Маржа': x['Маржа'].sum(),
        'Маржа (процентное соотношение)': x['Маржа'].sum() / x['СуммаСтроки'].sum(),
        'Количество чеков': x['НомерЗаказаНаСайте'].nunique(),
        'Количество товаров': x['Количество'].sum(),
        'Количество клиентов': x['Телефон_new'].nunique(),        
        'Средний чек': x['СуммаСтроки'].sum() / x['НомерЗаказаНаСайте'].nunique(),
        'Среднее количество товаров в чеке': x['Количество'].sum() / x['НомерЗаказаНаСайте'].nunique(),
        'Ценность клиента': x['Маржа'].sum() / x['Телефон_new'].nunique()
    }
    return pd.Series(result).round(3)
```

```python
df_without_delivery.groupby(['МесяцДатыЗаказа']).apply(summary_all)
```

```python
print('кол-во новых клиентов в 4 месяце',len(df_without_delivery.Телефон_new.unique())-len(df_without_delivery[df_without_delivery.МесяцДатыЗаказа==3].Телефон_new.unique()))
```

## Выкупленные товары

```python
def summary_buy(x):
    result = {
        'Выручка': x['СуммаСтроки'].sum(),
        'Абсолютная Маржа': x['Маржа'].sum(),
        'Маржа (процентное соотношение)': x['Маржа'].sum() / x['СуммаСтроки'].sum(),
        'Количество чеков': x['НомерЗаказаНаСайте'].nunique(),
        'Количество товаров': x['КоличествоПроданоКлиенту'].sum(),
        'Количество клиентов': x['Телефон_new'].nunique(),        
        'Средний чек': x['СуммаСтроки'].sum() / x['НомерЗаказаНаСайте'].nunique(),
        'Среднее количество товаров в чеке': x['КоличествоПроданоКлиенту'].sum() / x['НомерЗаказаНаСайте'].nunique(),
        'Ценность клиента': x['Маржа'].sum() / x['Телефон_new'].nunique()
    }
    return pd.Series(result).round(3)
```

```python
df_clean.groupby(['МесяцДатыЗаказа']).apply(summary_buy)
```
